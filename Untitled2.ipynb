```python
# combine a list of mixed values into a single string such that numeric values are squared , 
# non numeric values are reversed and all ouyputs are separated by :: 


lst = [2, "hi", 3, "abc"]
a = []
for x in lst:
    if type(x) == int or type(x) == float:
        a.append(str(x * x))
    else:
        a.append(x[::-1])
print (a)
op = "::".join(a)
print(op)
```

```python
# give two collection of values , compute common elements and non common elements without using any direct built-in shortcuts 

a = [1,2,3,4,5]
b = [4,5,6,7,8]

common = []
non_common = []

for x in a:
    if x in b:
        common.append(x)
    
for x in b:
    if x not in a:
        non_common.append(x)

print("Common:", common)
print("Non-common:", non_common)
```

```python
#safely remove entries from a mapping where keys are numeric string and values cannot be  converted to integer 

d = {"1": "100", "2": "abc", "3": "200", "123": "300"}

for k in list(d):
    if k.isdigit() and not d[k].isdigit():
        del d[k]

print(d)
```

```python
# design a program where different inputss trigger normal execution , error handling and guarenteed cleanup logic respectively

try:
    x = int(input("Enter numerator: "))
    y = int(input("Enter denominator: "))

    result = x / y              
    print("Result:", result)

except ZeroDivisionError:          
    print("Error: Cannot divide by zero!")

else:
    print("Result:", result)

finally:                          
    print("Finished")
```

```python
# from a list contaning numbers , numeric strings , empty strings and invalid entries , 
#extract valid integer , igonres other silently and compute minimum , maximum and avrage


lst = [10, "20", "", "abc", 5, "30", "xyz"]

valid_numbers = []

for x in lst:
    try:
        valid_numbers.append(int(x))
    except:
        pass 

if valid_numbers:
    minimum = min(valid_numbers)
    maximum = max(valid_numbers)
    average = sum(valid_numbers) / len(valid_numbers)
    print("Min:", minimum)
    print("Max:", maximum)
    print("Average:", average)
else:
    print("No valid numbers found")
```
